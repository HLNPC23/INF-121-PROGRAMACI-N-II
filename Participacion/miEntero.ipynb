```python
class miEntero:
    def __init__(self, valor):
        self.valor = valor

    # Método get
    def get(self):
        return self.valor

    # Métodos de instancia
    def esPar(self):
        return self.valor % 2 == 0

    def esImpar(self):
        return self.valor % 2 != 0

    def esPrimo(self):
        if self.valor < 2:
            return False
        for i in range(2, int(self.valor ** 0.5) + 1):
            if self.valor % i == 0:
                return False
        return True

    # Métodos estáticos con int
    @staticmethod
    def esParInt(n: int):
        return n % 2 == 0

    @staticmethod
    def esImparInt(n: int):
        return n % 2 != 0

    @staticmethod
    def esPrimoInt(n: int):
        if n < 2:
            return False
        for i in range(2, int(n ** 0.5) + 1):
            if n % i == 0:
                return False
        return True

    # Métodos estáticos con objeto miEntero
    @staticmethod
    def esParObj(obj):
        return obj.get() % 2 == 0

    @staticmethod
    def esImparObj(obj):
        return obj.get() % 2 != 0

    @staticmethod
    def esPrimoObj(obj):
        valor = obj.get()
        if valor < 2:
            return False
        for i in range(2, int(valor ** 0.5) + 1):
            if valor % i == 0:
                return False
        return True

    # Métodos equals
    def equalsInt(self, m: int):
        return self.valor == m

    def equalsObj(self, obj):
        return self.valor == obj.get()

    # parseInt
    @staticmethod
    def parseInt(chars):
        return int("".join(chars))
```

```python
n1 = miEntero(7)
print(n1.get())
print(n1.esPar())
print(n1.esImpar())
print(n1.esPrimo())

n2 = miEntero(10)
print(n2.get())
print(n2.esPar())
print(n2.esImpar())
print(n2.esPrimo())

print("Métodos estáticos con int")
print(miEntero.esParInt(6))
print(miEntero.esImparInt(5))
print(miEntero.esPrimoInt(7))

print("Métodos estáticos con objeto miEntero")
print(miEntero.esParObj(n1))
print(miEntero.esImparObj(n1))
print(miEntero.esPrimoObj(n1))
print(miEntero.esParObj(n2))
print(miEntero.esImparObj(n2))
print(miEntero.esPrimoObj(n2))

print("Métodos equals y parseInt")
n3 = miEntero(123)
print(n3.get())
print(n3.equalsInt(123))
print(n3.equalsObj(miEntero(123)))
print(miEntero.parseInt(['1','2','3']))
print(miEntero.parseInt("456"))
```
